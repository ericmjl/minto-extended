This is the notebook in which all post-processing of the pickled graph happens.

There are nodes that are unconnected to the 79 isolates of interest. We will remove them.

```python
import networkx as nx
import pandas as pd
import custom_funcs as cf
```

```python
G = nx.read_gpickle('../data/20160411_Minto_Run2 Final Graph.pkl')
cf.print_graph_data(G)
```

```python
# Read in the taxa of interest
toi = pd.read_csv('../data/Alaska_waterfowl_79viruses_metadata_20151223.csv')
toi
```

```python
subgraphs = nx.connected_component_subgraphs(G.to_undirected())

for g in subgraphs:
    has_toi = False
    for taxa in toi['Strain_name'].values:
        if g.has_node(taxa):
            has_toi = True
    
    if not has_toi:
        print('Removing {0} nodes...'.format(len(g.nodes())))
        G.remove_nodes_from(g.nodes())
```
